C'est très simple avec [la fonction `plot`](https://matplotlib.org/3.1.0/tutorials/introductory/pyplot.html) de la bibliothèque [Matplotlib](https://matplotlib.org/3.1.0/tutorials/), surtout si on utilise [NumPy](https://www.courspython.com/apprendre-numpy.html) pour manipuler des [intervalles de valeurs linéaires](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html). Il suffit d'utiliser la fonction [`fill_between`](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.fill_between.html).


Avant de l'utiliser pour afficher notre programme linéaire, testons ça sur un petit exemple. Mettons qu'on veuille représenter l'inéquation $2x_1 + x_2 \leq 4$



````python
import numpy as np
import matplotlib.pyplot as plt

x1_max = 5
x1 = np.linspace(0, x1_max) # on genere l'intervalle des x
x2 = 8 - 2*x1 # on genere l'intervalle des y

x2_max = np.amax(x2)

plt.plot(x1, x2)
plt.fill_between(x1, x2, x2_max, color='lightgrey')

# on dessine entre 0,0 et les extremas des fonctions
plt.xlim(0, x1_max)
plt.ylim(0, x2_max)

# on affiche
plt.show()
````

On a donc tout ce qu'il nous faut pour afficher les inéquations de nos contraintes. Traçons les droites en rouge, et les zones d'inégalités en gris.



````python
import numpy as np
import matplotlib.pyplot as plt

xmax = 20
x1 = np.arange(0, xmax)

# Contraintes
y1 = 8 - 2*x1
y2 = (7 - x1)/2
y3 = np.full_like(x1, 3) # ligne horizontale y = 3

# Valeur maximale utile du graphe
ymax = max(np.max(y1), np.max(y2), np.max(y3))

# Tracé des contraintes
plt.plot(x1, y1, label="2x + y ≤ 8", color='red')
plt.plot(x1, y2, label="x + 2y ≤ 7", color='blue')
plt.plot(x1, y3, label="y ≤ 3", color='green')

# Remplissage de la zone faisable (en dessous des trois contraintes)
plt.fill_between(x1, 0, np.minimum(np.minimum(y1, y2), y3), color='lightgrey')

# Axes et limites
plt.ylim(0, ymax)
plt.xlim(0, xmax)

plt.xlabel("x (kg de yaourt A)")
plt.ylabel("y (kg de yaourt B)")
plt.title("Zone faisable – Problème de fabrication de yaourts")

plt.legend()
plt.gcf().set_size_inches(18, 10)
plt.show()
````

Voici un tableau avec les pourcentages de chaque matière dans l'alliage et son coût par unité&nbsp;:

|**% zinc**     | 30  | 40  | 20  | 40  | 30  | 30  | 50  | 30  | 10  |
|**% étain**    | 50  | 10  | 50  | 30  | 40  | 10  | 10  | 60  | 80  |
|**Coût unit.** | 7.3 | 6.9 | 7.3 | 7.5 | 7.6 | 6.0 | 5.8 | 4.3 | 4.1 |

Le département de R&D recherche un « nouvel alliage » à prix minimum qui serait un mélange des 9 autres avec les contraintes d'avoir au maximum 30% de plomb, 30% de zinc et 40% d'étain.

## 2.1 Modélisation du problème

Commençons par modéliser le problème sous la forme d'un programme linéaire. Appliquez la méthode de l'exercice précédent.<br />

<em>À COMPLÉTER :

On note Xi la proportion de l’alliage i dans le mélange final (i = 1,...,9), avec Xi ≥ 0.

On cherche à déterminer les proportions des 9 alliages permettant de fabriquer un nouvel alliage à coût minimal.

Fonction objectif :

Minimiser le coût total :
Z = 7.3X1 + 6.9X2 + 7.3X3 + 7.5X4 + 7.6X5 + 6.0X6 + 5.8X7 + 4.3X8 + 4.1X9

Sous contraintes :

Contrainte sur le plomb (maximum 30%) :
0.20X1 + 0.50X2 + 0.30X3 + 0.30X4 + 0.30X5 + 0.60X6 + 0.40X7 + 0.10X8 + 0.10X9 ≤ 0.30

Contrainte sur le zinc (maximum 30%) :
0.30X1 + 0.40X2 + 0.20X3 + 0.40X4 + 0.30X5 + 0.30X6 + 0.50X7 + 0.30X8 + 0.10X9 ≤ 0.30

Contrainte sur l’étain (maximum 40%) :
0.50X1 + 0.10X2 + 0.50X3 + 0.30X4 + 0.40X5 + 0.10X6 + 0.10X7 + 0.60X8 + 0.80X9 ≤ 0.40

## 2.2 Résolution
Nous sommes en présence d'un problème à 9 variables. La méthode de résolution graphique ne peut pas être appliquée ici, nous allons nous tourner vers le solveur [PuLP](https://pythonhosted.org/PuLP/) qui calculera lui-même la solution, en exploitant la propriété qu'on a identifiée dans l'exercice précédent.


On va s'inspirer d'[un exemple basique](https://www.coin-or.org/PuLP/CaseStudies/a_blending_problem.html) pour implémenter et résoudre le modèle&nbsp;:



````python
import sys
!{sys.executable} -m pip install pulp
````

````python
from pulp import *
import numpy as np

# variables du problème
X1 = LpVariable("X_1", 0, None)
X2 = LpVariable("X_2", 0, None)
X3 = LpVariable("X_3", 0, None)
X4 = LpVariable("X_4", 0, None)
X5 = LpVariable("X_5", 0, None)
X6 = LpVariable("X_6", 0, None)
X7 = LpVariable("X_7", 0, None)
X8 = LpVariable("X_8", 0, None)
X9 = LpVariable("X_9", 0, None)



# probleme
prob = LpProblem("alliage", LpMinimize)

# objectif
prob += (
    7.3*X1 + 6.9*X2 + 7.3*X3 + 7.5*X4 + 7.6*X5
    + 6.0*X6 + 5.8*X7 + 4.3*X8 + 4.1*X9
)
                 
# contraintes
prob += X1+X2+X3+X4+X5+X6+X7+X8+X9 == 1

prob += (
    0.20*X1 + 0.50*X2 + 0.30*X3 + 0.30*X4 + 0.30*X5
    + 0.60*X6 + 0.40*X7 + 0.10*X8 + 0.10*X9 <= 0.30
)

prob += (
    0.30*X1 + 0.40*X2 + 0.20*X3 + 0.40*X4 + 0.30*X5
    + 0.30*X6 + 0.50*X7 + 0.30*X8 + 0.10*X9 <= 0.30
)

prob += (
    0.50*X1 + 0.10*X2 + 0.50*X3 + 0.30*X4 + 0.40*X5
    + 0.10*X6 + 0.10*X7 + 0.60*X8 + 0.80*X9 <= 0.40
)

prob.solve()
print(LpStatus[prob.status])
print("Min=", value(prob.objective))

# variables resultat
for v in prob.variables():
    print("%s=%.2f"%(v.name,v.varValue), end=', ')
````

Nous avons notre solution optimale&nbsp;! On peut même pousser le raffinement jusqu'à utiliser $\mathrm{\LaTeX}$ pour améliorer le rendu&nbsp;:

````python
from IPython.display import display, Latex

# variables resultat
result = ""
for v in prob.variables():
    valeur = round(v.varValue*100, 2)
    if valeur > 0:
        result += "$"+v.name+"="+str(valeur)+"\%$, "

display(Latex(result))
````

Une remarque&nbsp;: PuLP accepte aussi bien des problèmes de minimisation que de maximisation, et des contraintes d'infériorité que de supériorité. Mais tous les solveurs ne sont pas aussi souples, certains n'acceptent que programmes linéaires sous forme canonique pure, d'autres sous forme standard. Pourquoi&nbsp;? À votre avis, comment PuLP peut-il faire pour s'interfacer avec ces solveurs solveurs&nbsp;? 

<em>À COMPLÉTER : Certains solveurs ne savent résoudre les programmes linéaires que sous une forme mathématique précise, appelée forme canonique ou forme standard, car leurs algorithmes ont été conçus pour fonctionner uniquement avec ce type d’écriture. Par exemple, certains exigent une fonction objectif de minimisation et des contraintes uniquement sous forme d’égalités ou d’inégalités dans un seul sens. PuLP peut tout de même utiliser ces solveurs car il transforme automatiquement le problème saisi par l’utilisateur dans la forme attendue : il convertit une maximisation en minimisation, change le sens des inégalités si nécessaire, ajoute des variables d’écart pour transformer les inégalités en égalités, et adapte les variables afin de respecter les contraintes imposées par le solveur.</em>

Par ailleurs, l'algorithme du Simplexe n'est pas le seul sachant résoudre un programme linéaire. PuLP peut en utiliser un autre. Le connaissez-vous&nbsp;? Quels sont leurs principes généraux de fonctionnement, et leurs performances&nbsp;?

<em>À COMPLÉTER : Un autre type d’algorithme pouvant être utilisé par PuLP est la méthode des points intérieurs. Le principe du Simplexe est de parcourir les sommets de la région admissible en se déplaçant de sommet en sommet jusqu’à atteindre la meilleure solution possible. À l’inverse, les méthodes de points intérieurs explorent l’intérieur de la zone faisable et se dirigent progressivement vers la solution optimale sans suivre uniquement les bords. Le Simplexe est généralement très rapide pour les problèmes de petite ou moyenne taille, tandis que les méthodes de points intérieurs sont souvent plus performantes pour les très grands problèmes de programmation linéaire.</em>

Σ (poids_i * xi) ≤ 500

Et avec :
xi ∈ {0,1} pour tout i
</em>


<br />

Un algorithme naïf pour déterminer un résultat optimal serait une approche par force brute, qui teste toutes les solutions (admissibles ou non) et conserve la meilleure solution admissible. Dans le cas de notre étudiant, c'est une solution qui semble envisageable. Mais si l'étudiant supprime plusieurs Giga octets de jeux, et qu'il enrichit sa bibliothèque musicale, cette solution restera-t-elle viable&nbsp;? Qu'est-ce qui va devenir bloquant&nbsp;?

<em>À COMPLÉTER : 
Une approche par force brute devient rapidement impossible lorsque le nombre de titres augmente, car le nombre de combinaisons possibles croît de manière exponentielle (2^n).

Ce qui devient bloquant est donc la complexité combinatoire : tester toutes les solutions devient trop long en temps de calcul dès que la bibliothèque contient beaucoup d’éléments.
</em>


## 3.2 Résolution

Puisque nous avons modélisé le problème sous forme linéaire, pourquoi ne pas tester l'algorithme du Simplexe dessus&nbsp;? Reprenons les données issues du Smartphone de notre étudiant. Attention aux bornes de nos variables&nbsp;!

````python
# variables du problème
X1  = LpVariable("X_01", 0, 1)
X2  = LpVariable("X_02", 0, 1)
X3  = LpVariable("X_03", 0, 1)
X4  = LpVariable("X_04", 0, 1)
X5  = LpVariable("X_05", 0, 1)
X6  = LpVariable("X_06", 0, 1)
X7  = LpVariable("X_07", 0, 1)
X8  = LpVariable("X_08", 0, 1)
X9  = LpVariable("X_09", 0, 1)
X10 = LpVariable("X_10", 0, 1)


# probleme
prob = LpProblem("knapsack", LpMaximize)

# objectif
prob += (
    2*X1 + 5*X2 + 1*X3 + 2*X4 + 3*X5 +
    5*X6 + 2*X7 + 3*X8 + 1*X9 + 1*X10
)

# contrainte
prob += (
    40*X1 + 52*X2 + 27*X3 + 68*X4 + 80*X5 +
    62*X6 + 63*X7 + 53*X8 + 56*X9 + 59*X10 <= 500
)

prob.solve()
print(LpStatus[prob.status])
print("Valeur totale = ", value(prob.objective))

poids_total = 0.0
poids = [40, 52, 27, 68, 80, 62, 63, 53, 56, 59]
for idx, v in enumerate(prob.variables()):
    print("X%d=%f" % (idx+1, v.varValue), end=', ')
    poids_total += v.varValue*poids[idx]
print("\npoids=%f" % poids_total)
````

Ça vous parait correct&nbsp;? Pourtant, cela ne l'est pas&nbsp;! Calculez à la main le poids de la solution, en reprenant les valeurs dans le tableau. Qu'obtenez-vous&nbsp;? Pourquoi&nbsp;?

<em>À COMPLÉTER : Non, cette solution n’est pas correcte.

En recalculant le poids à la main à partir du tableau, on obtient :

40 + 52 + 27 + 68 + 80 + 62 + 63 + 53 = 445 Mo

Donc le poids total est bien 445 Mo et non 500 Mo.

La valeur affichée de 500 Mo vient du fait que la contrainte est une inégalité (≤ 500), et le solveur ne remplit pas forcément exactement toute la capacité.

Il respecte la contrainte, mais ne l’utilise pas entièrement car certaines combinaisons plus coûteuses ne sont pas rentables en termes de gain de notes.

</em>


Vérifions ça&nbsp;:

````python
poids_total=0.0
for idx, v in enumerate(prob.variables()):
    poids_total += v.varValue*poids[idx]
print("\npoids=%d"%poids_total)
````

Et là ça ne va plus du tout, La solution n'est pas admissible. On a vu qu'un arrondi inférieur produisait une solution qui, elle, est admissible. Mais avec une valeur de $X_9$ aussi proche de 1, ce n'est peut-être pas la solution optimale. Est-ce qu'un <a href="https://www.w3schools.com/python/ref_func_round.asp">arrondi au plus proche</a> marcherait mieux&nbsp;?

````python
poids_total=0.0
for idx, v in enumerate(prob.variables()):
    print("X%d=%f" % (idx+1, v.varValue), end=', ')
    poids_total += int(round(v.varValue))*poids[idx]
print("\npoids=%d" % poids_total)
````

Qu'observe-t-on&nbsp;?

<em>À COMPLÉTER : On observe que l’arrondi au plus proche ne garantit pas une solution admissible.

En effet, certaines variables proches de 1 dans la solution continue peuvent, une fois arrondies, violer la contrainte de capacité (500 Mo).

De plus, une solution arrondie peut devenir inférieure à une autre solution admissible existante, ce qui signifie qu’elle n’est pas forcément optimale.</em>

Parmi toutes les solutions générées, nous en avons une solution qui est admissible. Est-ce une bonne méthode pour trouver une solution optimale&nbsp;? Prenons cet exemple, dans lequel l'espace des solutions admissibles délimité par les deux contraintes est en blanc. Est-ce que, pour un problème _de maximisation_, arrondir une solution optimale réelle fonctionnerait&nbsp;?


<em>À COMPLÉTER : Non, ce n’est pas une bonne méthode pour obtenir une solution optimale.

Dans un problème de maximisation avec contraintes, arrondir une solution continue peut :
- violer les contraintes,
- ou éliminer des combinaisons pourtant optimales.

Ainsi, même si la solution semble proche de la frontière admissible, l’arrondi ne garantit ni faisabilité ni optimalité.
</em>

C'est finalement assez logique, puisque le problème de décision est $\mathcal{NP}$-Complet ([on le sait depuis 1972](https://fr.wikipedia.org/wiki/21_probl%C3%A8mes_NP-complets_de_Karp)). Le nombre exponentiel de solution, qu'on a démontré tout à l'heure, était-il un indice&nbsp;?

<em>À COMPLÉTER : 
Cela est cohérent avec le fait que le problème du sac à dos est NP-complet.

Le nombre exponentiel de solutions possibles est un indice clair de cette complexité.

Cela signifie qu’il n’existe pas d’algorithme connu permettant de trouver systématiquement la solution optimale en temps polynomial, ce qui explique pourquoi les méthodes naïves (force brute ou arrondis) ne sont pas fiables.
</em>
